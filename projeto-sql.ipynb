# 1.Breve descrição da base de dados:

Este projeto tem como função realizar uma análise de dados de crédito contidos em um arquivo CSV disponível nesse [link](https://raw.githubusercontent.com/andre-marcos-perez/ebac-course-utils/develop/dataset/credito.csv). Este projeto tem como intuito realizar a análise e exploração dos dados cotidos neste link.
Foi utilzado apenas uma fração dos dados contidos na base, com a finalidade de obter uma representação parcial do perfil dos clientes.

| Coluna  | Descrição |
| ------- | --------- |
| id      | Número da conta bancária |
| idade   | 26 a 73 anos |
| sexo    | F ou M |
| depedentes | 0 a 5 |
| escolaridade | Sem educação formal, ensino médio, graduação, mestrado ou doutorado |
| estado_civil | Solteiro, casado ou divorciado |
| salario_anual | Faixa do salário mensal multiplicado por 12 |
| tipo_cartao | Categoria do cartão: blue, silver, gold ou platinum |
| qtd_produtos | Quantidade de produtos contratados |
| iteracoes_12m | Quantidade de iterações com o cliente no último ano |
| meses_inativo_12m | Quantidade de meses que o cliente ficou inativo no último ano |
| limite_credito | Valor do limite do cartão de crédito |
| valor_transacoes_12m | Soma total do valor das transações no cartão de crédito no último ano |
| qtd_transacoes_12m | Quantidade total de transações no cartão de crédito no último ano |

A tabela foi criada no AWS Athena com o auxílio do S3 Bucket.

# 3. Análise de dados

Vamos observar a disposição das faixas salariais desconsiderando dados ausentes:

##### query: 

select count (*) as quantidade, salario_anual FROM credito 
where salario_anual != 'na'
group by salario_anual

![](https://github.com/MarcosFernandesRocha/Projeto-analise-credito-SQL/blob/main/Query%209.jpeg?raw=true)

![](https://github.com/MarcosFernandesRocha/Projeto-analise-credito-SQL/blob/main/Gr%C3%A1fico%201.jpeg?raw=true)

Referente a disposição por sexo:

##### query: select count(*) as quantidade, sexo from credito group by sexo

![](https://github.com/MarcosFernandesRocha/Projeto-analise-credito-SQL/blob/main/Query%2010.jpeg?raw=true)

![](https://github.com/MarcosFernandesRocha/Projeto-analise-credito-SQL/blob/main/Gr%C3%A1fico%202.jpeg?raw=true)

A maior parte dos clientes analisados da base são do sexo masculino. o Gráfico de pizza facilita a visualização das informações.

##### Com a finalidade de conhecer melhor meus clientes e direcionar campanhas de produtos e serviços, vamos avaliar os perfis de idades
##### Query: select avg(idade) as media_idade, min(idade) as minimo_idade, max(idade) as maximo_idade, sexo from credito group by sexo

![](https://github.com/MarcosFernandesRocha/Projeto-analise-credito-SQL/blob/main/Query%2011.jpeg?raw=true)

Avaliando as colunas de idade por média por sexo, idade mínima por sexo e idade máxima por sexo podemos notar que não existe uma diferença relevante. Pelo contrário existe uma similaridade no perfil de clientes de acordo com a amostra da base que foi extraída.

##### Quais as características dos clientes que possuem os maiores créditos?

##### Query: 
select max(limite_credito) as limite_credito, escolaridade, tipo_cartao, sexo from credito  
where escolaridade != 'na' and tipo_cartao != 'na'  
group by escolaridade, tipo_cartao, sexo  
order by limite_credito desc  
limit 10

![](https://github.com/MarcosFernandesRocha/Projeto-analise-credito-SQL/blob/main/Query%2012.jpeg?raw=true)

Não é possível identificar uma relação entre escolaridade e tipo de cartão com os maiores limites de crédito, tendo em vista que o maior limite de crédito desta base se econtra com um cliente sem educação formal. Nessa amostra, os tipos de cartão variam entre todas as categorias disponíveis. Entretanto, observa-se nessa consulta uma predominância de clientes do sexo masculino.

##### Quais as características dos clientes que possuem os menores créditos?
##### Query:
select max(limite_credito) as limite_credito, escolaridade, tipo_cartao, sexo from credito  
where escolaridade != 'na' and tipo_cartao != 'na'  
group by escolaridade, tipo_cartao, sexo  
order by limite_credito asc  
limit 10

![](https://github.com/MarcosFernandesRocha/Projeto-analise-credito-SQL/blob/main/Query%2013.jpeg?raw=true)

Mais uma vez a escolaridade e tipos de cartão não demonstram uma relação direta com os limites de crédito. Porém novamente, podemos perceber que a coluna de sexo possui uma relação com a faixa de limite de crédito. Agora o sexo feminino se mostra mais presente basta observarmos os menores indices de limietes de crétito.

##### Será que as mulheres gastam mais que os homens?

###### Query: 
select max(valor_transacoes_12m) as maior_valor_gasto, avg(valor_transacoes_12m) as media_valor_gasto, min(valor_transacoes_12m) as min_valor_gasto, sexo
from credito  
group by sexo

![](https://github.com/MarcosFernandesRocha/Projeto-analise-credito-SQL/blob/main/Query%2014.JPG?raw=true)

##### O salário tem algum impacto no limite de crédito?
##### Query:
select avg(valor_transacoes_12m) as media_valor_transacoes, avg(limite_credito) as media_limite, sexo , salario_anual
from credito  
where salario_anual != 'na'  
group by sexo, salario_anual  
order by avg(limite_credito) desc



![](https://github.com/MarcosFernandesRocha/Projeto-analise-credito-SQL/blob/main/Query%2016.JPG?raw=true)

As pessoas com menores faixas salariais também possuem menor limites de crédito.

# 4. Conclusão 

- A maior parte dos clientes possuem renda atpe 40k
- A maior parte dos clinetes é do sexo masculino
- A escolaridade não tem impacto direto no limite de crédito
- Os clientes com menores limites são do sexo feminino
- Os clietes com maiores limites são do sexo masculino
- Dentre os menores limites não há presença de cartão platinum
- A faixa salarial impacta diretamente no limite de crédito
- Esta recorte da base de dados não apresentou clientes do sexo feminino com salário superior a 60k

Embora esteja sendo apresetado apenas um recorte da base geral de dados, é possível associar essa análise e exploração à realizade. A disponibilidade de maiores limites de crédito costuma estar atrelada ao salário e consequentemente ao comportamento do cliente. Sobre os clientes do sexo masculino possuírem maiores salários e limites de crédito, é possível levar em conta os fatores culturais da sociedade e trazer para uma discussão do cenário real.
